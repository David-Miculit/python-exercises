```python
vowels = ['a','e','i','o','u']
def analyze_word(word):
    length = len(word)
    vowels_nr = 0
    uppercase_word = word.upper()
    for i in word:
        if i in vowels:
            vowels_nr += 1
    return length, vowels_nr, uppercase_word

analyze_word("Morbius")
```

```python
nr_list = [1,2,4,6,8]

def number_stats(numbers):
    if (len(numbers) == 0):
        return "No numbers given"
    odd_nr = 0
    for i in numbers:
        if type(i) == str:
            return "Pass a list with only numbers"
        elif(i%2!=0):
            odd_nr += 1
    return min(numbers), max(numbers), sum(numbers)/len(numbers), odd_nr

number_stats(nr_list)
```

```python
list1 = [(1, "apple"), (3, "orange"), (5, "banana")]
list2 = [(2, "grape"), (4, "kiwi"), (6, "melon")]

def sorted_criteria(list):
    return list[0]

def sort_lists(list1, list2):
    merged_list = list1 + list2
    sorted_list = sorted(merged_list, key=sorted_criteria)
    return sorted_list

sort_lists(list1, list2)
```

```python
people = {
    "Alice": {"age": 30, "city": "New York"},
    "Bob": {"age": 25, "city": "Los Angeles"},
    "Charlie": {"age": 35, "city": "Chicago"},
    "John": {"age": 20, "city": "New York"}
}

def oldest_person(people):
    maxx=0
    for key in people:
        if(people[key]["age"] > maxx):
            maxx = people[key]['age']
            oldest = key
    return oldest

def new_yorkers(people):
    new_yorkers = []
    for key in people:
        if(people[key]["city"] == "New York"):
            new_yorkers.append({key: people[key]})
    return new_yorkers
      
print("Oldest person:", oldest_person(people))
print("People that live in NY:", new_yorkers(people))
```
