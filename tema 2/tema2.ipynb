```python
nr = int(input("Choose a number: "))
print(nr)

if nr > 0 and nr%2==0:
    print("even positive nr")
elif nr > 0 and nr%2!=0:
    print("odd positive nr")
elif nr < 0 and nr%2==0:
    print("even negative nr")
elif nr < 0 and nr%2!=0:
    print("odd negative nr")
else:
    print("number is 0")
```

```python
vocale = ['a','e','i','o','u']
litera = input("Choose a letter: ")
if litera in vocale:
    print("litera e vocala")
elif len(litera)==1:
    print(f"{litera} e consoana")
else:
    print("nu e litera")
```

```python
age = int(input("Choose your age: "))
if age < 18 or age > 65:
    print(f"Eligible for discount at {age} y/o")
else:
    print(f"Ineligible for discount at {age} y/o")
```

```python
password = input("Type your password: ")
print(f"Password: {password}")
if len(password) < 8 or password.isalpha() == True or password.islower() == True:
    print("Password is too weak")
else:
    print("Password is strong")
```

```python
vowels = ['a','e','i','o','u']
vowels_found = []
word = input("Choose a word: ")
counter = 0
for i in word:
    if i in vowels:
        counter += 1
        vowels_found.append(i)
print(f"{counter} vowels in {word}")
print(vowels_found)
```
